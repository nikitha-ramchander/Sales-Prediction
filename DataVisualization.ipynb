```
# Imports 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Load Data
df = pd.read_csv('/content/drive/MyDrive/Coding Dojo/Project/salespredictions.csv')
df.head()
```

```
df_1 = df.drop(columns = ['Outlet_Identifier', 'Outlet_Location_Type', 'Outlet_Size'])
df_1.head(2)
```

#Data Visualization 
>Built several data visualizations that help better understand the data.

```
# Checked to see if it is a pandas series
df_1['Outlet_Type'].value_counts().index
```

#1st Graph

```
outlet_type = df_1.groupby(['Outlet_Type'])['Item_Outlet_Sales'].sum().sort_values(ascending = False)
```

```
plt.figure(figsize=(10,5));
plt.bar(outlet_type.index, outlet_type.values);
#plt.bar(df_1.groupby(['Outlet_Type'])['Item_Outlet_Sales'].sum().index, df_1.groupby(['Outlet_Type'])['Item_Outlet_Sales'].sum().values);
plt.xticks(fontsize = 12);
plt.yticks(fontsize = 12);
plt.xlabel('Types of Outlets', fontsize = 16);
plt.ylabel('Total Sales', fontsize = 16);
plt.title('Total Sales for Each Type of Outlet', fontsize = 16);
```

#2nd Graph

```
establishment = df_1.groupby(['Outlet_Establishment_Year'])['Item_Outlet_Sales'].sum()
```

```
plt.figure(figsize=(10,5));
plt.plot(establishment.index, establishment.values);
plt.xticks(fontsize = 12);
plt.yticks(fontsize = 12);
plt.xlabel('Year of Outlet Establishment', fontsize = 16);
plt.ylabel('Total Sales', fontsize = 16);
plt.title('Total Sales for Each Year of Establishment', fontsize = 16);
```

#3rd Graph

```
 item_type = df_1.groupby(['Item_Type'])['Item_Outlet_Sales'].sum().sort_values()
```

```
plt.figure(figsize=(10,5));
plt.barh(item_type.index, item_type.values)
plt.xticks(fontsize = 12);
plt.yticks(fontsize = 12);
plt.xlabel('Total Sales', fontsize = 16)
plt.ylabel('Types of Items', fontsize = 16)
plt.title('Total Sales for Each Type of Item', fontsize = 16)
```
